## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def line_slope (x1, y1, x2, y2):
    
    if(x2 - x1)!= 0:
        slope = (y2 - y1)/(x2 - x1)
    else:
        slope = 0
    
    return slope


def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """

    sum_slope_left = 0
    sum_slope_right = 0
    sum_yintcpt_left = 0
    sum_yintcpt_right = 0
    num_left_lines = 0 #number of left lane markers 
    num_right_lines = 0 #number of right lane markers
    abs_min_slope = 0.15 #absolute value for the min value of slope allowed
    abs_max_slope = 0.85 #absolute value for the max value of slope allowed

    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = line_slope(x1, y1, x2, y2)
            if slope > 0 and slope > abs_min_slope and slope < abs_max_slope:
                #Positive slope indicates it belongs to the left lane marker
                #Calculate the slope and y intercept and sum them up respectively (used to later find the average slope and y intercept)
                sum_slope_left = sum_slope_left + slope
                sum_yintcpt_left = sum_yintcpt_left + y1 - (slope * x1)
                num_left_lines = num_left_lines + 1

            elif slope < 0 and abs(slope) > abs_min_slope and abs(slope) < abs_max_slope:
                #Negative slope indicates it belongs to the right lane marker
                ##Calculate the slope and y intercept and sum them up respectively (used to later find the average slope and y intercept)
                sum_slope_right = sum_slope_right + slope
                sum_yintcpt_right = sum_yintcpt_right + y1 - (slope * x1)
                num_right_lines = num_right_lines + 1
    
    #Fix the y co-ord for the top and bottom of the region of interest and then calculate the corresponding 
    # values of the x-coordinate using the average slope and average y interept values
    bottom_y = 540
    top_y = 321
    
    if (num_left_lines > 0):  
        
        #Calculate averge slope and y intercept of left lane lines
        avg_slope_left = sum_slope_left/num_left_lines 
        avg_yintcpt_left = sum_yintcpt_left/num_left_lines
        
        #Calculate the x-coordinate of the bottom left and top left extremities of the region of interest
        bottom_left_x = int((bottom_y - avg_yintcpt_left)/avg_slope_left)
        top_left_x = int((top_y - avg_yintcpt_left)/avg_slope_left)
        
        #Draw the left lane line
        cv2.line(img, (bottom_left_x, bottom_y), (top_left_x, top_y), color, thickness)
        
    if (num_right_lines > 0):
        
        #Calculate averge slope and y intercept of right lane lines
        avg_slope_right = sum_slope_right/num_right_lines 
        avg_yintcpt_right = sum_yintcpt_right/num_right_lines
        
        #Calculate the x-coordinate of the bottom right and top right extremities of the region of interest
        bottom_right_x = int((bottom_y - avg_yintcpt_right)/avg_slope_right)
        top_right_x = int((top_y - avg_yintcpt_right)/avg_slope_right)
        
        #Draw the right lane line
        cv2.line(img, (bottom_right_x, bottom_y), (top_right_x, top_y), color, thickness)  
    
    

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.
def find_lanes(img):
    # Pipeline to detect the lanes lanes in the image
    
    #Step: Make a copy of the input image
    img_in = np.copy(img)
    
    #The copy of the input image 'img_in' will be processed in the lane finding pipeline
    
    #Step: Convert to grascale
    gray_img = grayscale(img_in)    
    
    #Step: Gaussian blur on the grascale image
    kernel_size = 13
    blur_gray = gaussian_blur(gray_img, kernel_size)
    
    #Step: Apply the Canny function on the blurred image to find the edges
    canny_low_thresh = 30
    canny_high_thresh = 100
    edges_img = canny(blur_gray,canny_low_thresh, canny_high_thresh)
    
    #Step: Select the region of interest (roi) and mask out the rest of the image
    #Here a triangular region which contains the bottom half of the picture is considered as the roi
    imshape = edges_img.shape
    vertices = np.array([[(0,540),(470,321),(510,321),(960,540)]], dtype=np.int32)
    masked_edges_img = region_of_interest(edges_img, vertices)
    
    #Step: Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 22    # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 18 #minimum number of pixels making up a line
    max_line_gap = 1    # maximum gap in pixels between connectable line segments
    
    # Step: Run Hough on edge detected image
    line_image = hough_lines(masked_edges_img, rho, theta, threshold, min_line_length, max_line_gap)
    
    #Step: 
    #Overlay the hough lines image on the original image
    output_img = weighted_img(line_image, img_in)
    
    return output_img
````
